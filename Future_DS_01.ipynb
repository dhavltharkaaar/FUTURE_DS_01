Lets Rename the columns as per our requirements

```python
# Rename columns
df.rename(columns={
    'InvoiceNo': 'invoice_no',
    'InvoiceDate': 'order_date',
    'Description': 'description',
    'CustomerID': 'customer_id',
    'Quantity': 'quantity',
    'UnitPrice': 'price'
}, inplace=True)
```

```python
df.columns # display the name of the columns
```

```python
df.info()
```

 order_date column is in object which is a wrong datatype we will change it to datetime

```python
# Convert types
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df.info()
```

We can see only two columns description and customer_id as null value in which customer_id has max null values

```python
# Calculate percentage of nulls in customer_id 
cid = df['customer_id'].isnull().sum() / len(df) * 100 
d = df['description'].isnull().sum() / len(df) * 100
print(f"Percentage of missing customer_id: {cid:.2f}%")
print(f"Percentage of missing description: {d:.2f}%")
```

Before we proceed we can also see that there are many negative values into the dataset in column Quantity and price which are invalid values we can't have those value for analysis so we will make sure only valid vlaues are counted into our dataset

```python
# Keep only positive quantity and price
df = df[(df['quantity'] > 0) & (df['price'] > 0)]
```

```python
df.info()
```

```python
df.isnull().sum() # checks whether are there any null values in dataset
```

After dropping the invalid negative values from the data we still have null values in customer_id

```python
custid = df['customer_id'].isnull().sum() / len(df) * 100
print(f"Percentage of missing customer_id: {custid:.2f}%")
```

```python
# Drop rows with missing customer_id  
df.dropna(subset=['customer_id'], inplace=True)
```
